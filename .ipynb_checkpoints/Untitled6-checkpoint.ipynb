```python
import scipy
import numpy as np
```

```python
N=15
def to_bin(x):
    return np.array([int(x) for x in '0'*(N-len(bin(x)[2:]))+bin(x)[2:]])

def loss(thetas):
    l = 0
    for n in range(2**N-1):
        l+= abs( (2*n-2**N)/(2*2**N) - np.sin(to_bin(n).dot(thetas[1:])+thetas[0]))**2
    return l
res = scipy.optimize.minimize(loss, x0 = np.zeros(N+1),bounds=[[-np.pi,np.pi]]*(N+1),options={'maxiter':10**10})
```

```python
res.fun
```

```python
loss(res.x)
```

```python
loss([-np.pi/6]+[np.arcsin(1/2**i) for i in range(1,N+1)])
```
