```python
import scipy
import numpy as np
```

```python
N=10
def to_bin(x):
    return np.array([int(x) for x in '0'*(N-len(bin(x)[2:]))+bin(x)[2:]])

def loss(thetas):
    l = 0
    single_thetas = thetas[1:-N*(N-1)//2]
    double_thetas = thetas[-N*(N-1)//2:]
    
    for n in range(2**N-1):
        phase = thetas[0]
        b = to_bin(n)
        for i in range(N):
            if(b[i]==1):
                phase += single_thetas[i]
            for j in range(i+1,N):
                if(b[i]==1 and b[j]==1):
                    phase += double_thetas[(N-1+N-i)*i//2+j-i-1]
        l+= abs( (2*n-2**N)/(2*2**N) - np.sin(phase))**2
    return l
res = scipy.optimize.minimize(loss, x0 = np.zeros(N*(N-1)//2+N+1),bounds=[[-np.pi,np.pi]]*(N*(N-1)//2+N+1),options={'maxiter':10**10})
res2 = scipy.optimize.minimize(lambda x: loss(np.concatenate((x,[0]*(N*(N-1)//2)))), x0 = np.zeros(N+1),bounds=[[-np.pi,np.pi]]*(N+1),options={'maxiter':10**10})
```

```python
def sample(thetas):
    y = []
    single_thetas = thetas[1:-N*(N-1)//2]
    double_thetas = thetas[-N*(N-1)//2:]
    for n in range(2**N-1):
        phase = thetas[0]
        b = to_bin(n)
        for i in range(N):
            if(b[i]==1):
                phase += single_thetas[i]
            for j in range(i+1,N):
                if(b[i]==1 and b[j]==1):
                    phase += double_thetas[(N-1+N-i)*i//2+j-i-1]
        y.append(np.sin(phase))
    return y
```

```python
import matplotlib.pyplot as plt
t1 = [-np.pi/6]+[np.arcsin(1/2**i) for i in range(1,N+1)]+[0]*(N*(N-1)//2)
t2 = res.x
t3 = np.concatenate((res2.x,[0]*(N*(N-1)//2)))
x = np.array(range(0,2**N-1))
y = np.array(2*np.array(range(0,2**N-1))-2**N)/(2*2**N)
y1 = sample(t1)
y2 = sample(t2)
y3 = sample(t3)
#plt.plot(x,y)
plt.plot(x,abs(y1-y))
plt.plot(x,abs(y2-y))
plt.plot(x,abs(y3-y),'--')
```
